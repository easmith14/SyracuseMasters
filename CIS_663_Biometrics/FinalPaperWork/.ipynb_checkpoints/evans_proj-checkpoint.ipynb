```python
# Import and data readin
import numpy as np
import matplotlib.pyplot as plt

from numpy import genfromtxt

raw_data = genfromtxt('TrainingData_Raw.csv', delimiter=',')

# fix nan
raw_data[0,0]=1.0

train_data = np.vstack((raw_data[0:20,:],raw_data[30:50,:],raw_data[60:80,:],raw_data[90:110,:]))

num_samples = train_data.shape[1]-1
num_chips= 4
num_readings_per_chip = int(train_data.shape[0]/num_chips)
```

```python
# Take SVD of simulated dataset
U,Sigma,Vh = np.linalg.svd(train_data[:,1:])
# Look at sing val array

plt.plot(Sigma)
plt.xlabel('Singular value index')
plt.show()

plt.plot(Sigma[:10])
plt.xlabel('First few Singular value index')
plt.show()
```

```python
# what do sig vecs do
for sig in range(3):
    plt.plot(Vh[sig,:],label=str(sig))
plt.legend()
plt.show()
```

```python
# compile fingerprints
num_sig_vals = 10
fingerprints= train_data[:,1:]@Vh[:num_sig_vals,:].T
assert fingerprints.shape[0] == num_readings_per_chip*num_chips

colors = ['b','g','orange','r']

# view average fingerprint per chip
for ii in range(num_chips):
    chip_fingerprints = fingerprints[ii*num_readings_per_chip:(ii+1)*num_readings_per_chip,:]
    plt.plot(np.mean(chip_fingerprints,axis=0),label='fingerprints for chip '+str(ii+1))
    plt.fill_between(np.arange(num_sig_vals),
                    np.mean(chip_fingerprints,axis=0)-np.std(chip_fingerprints,axis=0),
                     np.mean(chip_fingerprints,axis=0)+np.std(chip_fingerprints,axis=0),
                     alpha = 0.5,
                     label='fingerprints for chip '+str(ii+1))
    
plt.legend()
plt.show()
```
